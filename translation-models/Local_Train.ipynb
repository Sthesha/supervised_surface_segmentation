```python
import importlib
import os
import sys
import pandas as pd
import pandas as pd
from pathlib import Path
from train import train_model
from pathlib import Path
import torch
import torch.nn as nn
from train import get_model, get_ds, run_validation, evaluate
import pandas as pd
# train_model(config)

from pathlib import Path
from seq2seq_train import Seq2SeqTrainer
def load_config(config_name):
    try:
        module = importlib.import_module(f"config_files.{config_name}")
        return module.config
    except ImportError:
        raise ValueError(f"Configuration '{config_name}' not found.")

def list_configs():
    configs_dir = "./config_files"
    config_files = [f[:-3] for f in os.listdir(configs_dir) if f.endswith('.py') and f != '__init__.py']
    return config_files

def load_all_configs():
    all_configs = {}
    for config_name in list_configs():
        all_configs[config_name] = load_config(config_name)
    return all_configs

def get_config_by_name(config_name):
    try:
        return load_config(config_name)
    except ValueError as e:
        print(f"Error: {e}")
        print("Available configurations:")
        for cfg in list_configs():
            print(f"  - {cfg}")
        return None

configs = list_configs()
configs
```

```python
import importlib
import os
import sys

def load_config(config_name):
    try:
        module = importlib.import_module(f"config_files.{config_name}")
        return module.config
    except ImportError:
        raise ValueError(f"Configuration '{config_name}' not found.")

def list_configs():
    configs_dir = "./config_files"
    config_files = [f[:-3] for f in os.listdir(configs_dir) if f.endswith('.py') and f != '__init__.py']
    return config_files

def load_all_configs():
    all_configs = {}
    for config_name in list_configs():
        all_configs[config_name] = load_config(config_name)
    return all_configs

def get_config_by_name(config_name):
    try:
        return load_config(config_name)
    except ValueError as e:
        print(f"Error: {e}")
        print("Available configurations:")
        for cfg in list_configs():
            print(f"  - {cfg}")
        return None

configs = list_configs()
configs
# ['nllb_eng_zulu_tok_tok',
#  'nllb_xhosa_zulu_syl_syl',
#  'nllb_eng_zulu_tok_syl',
#  'nllb_eng_zulu_tok_seg_one',
#  'bible_eng_zulu_tok_seg_two',
#  'nllb_xhosa_zulu_tok_tok',
#  'bible_eng_zulu_tok_seg_two_v2',
#  'bible_xhosa_zulu_syl_syl',
#  'nllb_eng_zulu_tok_seg_two_V2',
#  'bible_xhosa_zulu_tok_tok',
#  'bible_eng_zulu_tok_syl',
#  'bible_eng_zulu_tok_tok']
nllb_xhosa_zulu_tok_tok_v2
```

```python
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

def plot_losses(log_dir):
    # Load TensorBoard logs
    event_acc = EventAccumulator(log_dir)
    event_acc.Reload()

    # Extract loss values
    train_loss = event_acc.Scalars('train loss')
    val_loss = event_acc.Scalars('val loss')

    # Extract values and steps for plotting
    train_steps = [x.step for x in train_loss]
    train_values = [x.value for x in train_loss]

    val_steps = [x.step for x in val_loss]
    val_values = [x.value for x in val_loss]

    # Plot the losses
    plt.figure(figsize=(10, 5))
    plt.plot(train_steps, train_values, label='Train loss')
    plt.plot(val_steps, val_values, label='Validation loss')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.title('Train and Validation Loss')
    plt.legend()
    plt.show()

# Provide the path to the directory where TensorBoard logs are stored
log_dir = 'models/nllb_eng_zulu_tok_seg_two_V2/tensor_data'
plot_losses(log_dir)


# config = get_config_by_name("nllb_eng_zulu_tok_tok")

# config2 = get_config_by_name("nllb_eng_zulu_tok_tok_v2")
```
